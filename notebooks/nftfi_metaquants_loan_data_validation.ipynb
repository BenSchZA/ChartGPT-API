```python
import setup
import pandas as pd
import plotly.express as px

from data.utils import format_bigquery_column_names
```

```python
df_mq_missing_data = pd.read_csv('data/metaquants_loans.csv')
df_mq = pd.read_csv('data/nft_finance_p2p.csv')
df_mq = df_mq.query('protocol == "nftfi"')
df_mq = df_mq.rename(columns={'block_timestamp': 'date'})
df_mq.head()
```

```python
df_mq.sort_values('date')
```

```python
df_mq_missing_data.sort_values('block_timestamp')
```

```python
df_mq.columns
```

```python
assert df_mq.protocol.unique() == 'nftfi'
```

```python
df_nftfi = pd.read_csv('data/nftfi_loans.csv')
df_nftfi = format_bigquery_column_names(df_nftfi)
df_nftfi = df_nftfi.rename(columns={'loan_date': 'date'})
df_nftfi.head()
```

```python
df_nftfi.columns
```

```python
# Filter date ranges
df_mq['date'] = pd.to_datetime(df_mq.date)
df_mq.sort_values('date', inplace=True)
df_nftfi['date'] = pd.to_datetime(df_nftfi.date)
df_nftfi.sort_values('date', inplace=True)

df_mq_min_date = df_mq.date.min()
df_mq_max_date = df_mq.date.max()

display(f'{df_mq_min_date=}')
display(f'{df_mq_max_date=}')

df_nftfi_min_date = df_nftfi.date.min()
df_nftfi_max_date = df_nftfi.date.max()

display(f'{df_nftfi_min_date=}')
display(f'{df_nftfi_max_date=}')

min_date = max(df_mq_min_date, df_nftfi_min_date)
max_date = min(df_mq_max_date, df_nftfi_max_date)

display(f'{min_date=}')
display(f'{max_date=}')

df_mq = df_mq[(df_mq.date >= min_date) & (df_mq.date <= max_date)]
df_nftfi = df_nftfi[(df_nftfi.date >= min_date) & (df_nftfi.date <= max_date)]
```

```python
assert df_mq.transaction_hash.is_unique
```

```python
df_mq_duplicates = df_mq[df_mq.duplicated(keep=False)]
df_mq_duplicates
```

```python
# There are a limited number of duplicate MetaQuants loans by TX hash,
# let's just drop them for now and address with MetaQuants later
df_mq.drop_duplicates(inplace=True)
```

```python
assert df_mq.transaction_hash.is_unique
# DataQuants NFTfi loan IDs are unique because they make V1 loans negative
assert df_mq.loan_id.is_unique
```

```python
assert df_nftfi.transaction_hash.is_unique
# NFTfi loan IDs are not unique because V1 and V2 overlap
# Instead, loan_id should be used...
assert df_nftfi.id.is_unique
```

```python
df_nftfi.query(f'id == {df_nftfi[df_nftfi.id.duplicated(keep=False)].id.iloc[0]}')[['loan_id', 'id']]
```

```python
# Number of transactions that are in MetaQuants dataset and not in NFTfi dataset
difference_mq_nftfi = set(df_mq.transaction_hash).difference(set(df_nftfi.transaction_hash))
len(difference_mq_nftfi)
```

```python
# Number of transactions that are in NFTfi dataset and not in MetaQuants dataset
difference_nftfi_mq = list(set(df_nftfi.transaction_hash).difference(set(df_mq.transaction_hash)))
len(difference_nftfi_mq)
```

```python
# Number of transactions that are in both the NFTfi dataset and the MetaQuants dataset
intersection = list(set(df_mq.transaction_hash).intersection(set(df_nftfi.transaction_hash)))
len(intersection)
```

```python
# Find the intersecting rows in MetaQuants dataset
df_mq.query(f'transaction_hash in {intersection}').head()
```

```python
# Find the intersecting rows in NFTfi dataset
df_nftfi_intersecting = df_nftfi.query(f'transaction_hash in {intersection}')
df_nftfi_intersecting.head()
```

```python
# Find the different rows in NFTfi dataset
df_nftfi_difference = df_nftfi.query(f'transaction_hash in {difference_nftfi_mq}')
df_nftfi_difference.head()
```

```python
# See if there is a pattern in the rows that are different
for col in df_nftfi_difference:
    print(col)
    print(df_nftfi_difference[col].unique())
```

```python
# Create unique version of loan ID to match MetaQuants convention
df_nftfi['id_unique'] = df_nftfi['id']
df_nftfi.loc[df_nftfi.loan_contract.str.contains('v1'), 'id_unique'] = -df_nftfi['id']
assert df_nftfi['id_unique'].is_unique
df_nftfi['id_unique']
```

```python
# Number of transactions that are in MetaQuants dataset and not in NFTfi dataset
difference_mq_nftfi_by_id = set(df_mq.loan_id).difference(set(df_nftfi.id_unique))
assert len(difference_mq_nftfi_by_id) == len(difference_mq_nftfi)
len(difference_mq_nftfi_by_id)
```

```python
# Number of transactions that are in NFTfi dataset and not in MetaQuants dataset
difference_nftfi_mq_by_id = list(set(df_nftfi.id_unique).difference(set(df_mq.loan_id)))
assert len(difference_nftfi_mq_by_id) == len(difference_nftfi_mq)
len(difference_nftfi_mq_by_id)
```

```python
# Number of transactions that are in both the NFTfi dataset and the MetaQuants dataset
intersection_by_id = list(set(df_mq.loan_id).intersection(set(df_nftfi.id_unique)))
assert len(intersection_by_id) == len(intersection)
len(intersection_by_id)
```

```python
# Because both the intersection and difference by TX hash and ID are the same,
# we can assert that it's not a case of the NFTfi dataset having multiple
# transactions per loan (for example on repaid, liquidate, etc.)
```
